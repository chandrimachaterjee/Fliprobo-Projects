```python
def dis_bar(x,y):
    
    df.groupby([x,y]).size().unstack(level=-1).plot(kind='bar', figsize=(35,10))
    plt.xlabel(x,fontsize= 25)
    plt.ylabel('count',fontsize= 25)
    plt.legend(loc=0,fontsize= 25)
    plt.xticks(fontsize=30)
    plt.yticks(fontsize=30)
    plt.title("{X} Vs {Y}".format(X=x,Y=y),fontsize = 50)
    plt.show()
```

```python
#Separating categorical and numerical columns

cat_df,num_df=[],[]
for i in df:
    if df[i].dtypes=='O':
        cat_df.append(i)
    elif (df[i].dtypes=='int') | (df[i].dtypes=='float') | (df[i].dtypes=='int64'):
        num_df.append(i)
    else: print('Unknown >',i)
    
print('Total number of features:',df.shape[1])
print('Number of Categorical features:',len(cat_df))
print('Number of Numerical features:',len(num_df))
```

```python
for i in cat_df:
    f=plt.figure(figsize=(12,4))
    f.add_subplot(1,1,1)
    sns.countplot(df[i])
```
